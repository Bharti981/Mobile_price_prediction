Handle outliers in column with IQR method by capping

```python
def iqr_technique(DFcolumn):
    Q1 = np.percentile(DFcolumn, 25)
    Q3 = np.percentile(DFcolumn, 75)
    IQR = Q3 - Q1
    lower_range = Q1 - (1.5 * IQR)
    upper_range = Q3 + (1.5 * IQR)
    return lower_range, upper_range
```

```python
lower_limit_iqr,upper_limit_iqr = iqr_technique(df['Price'])
```

```python
df['Price'] = np.where(df['Price']>upper_limit_iqr,upper_limit_iqr,np.where(df['Price']<lower_limit_iqr,lower_limit_iqr,df['Price']))
```

# Feature creation

```python
def concat_colums(x,y):
    if x>0 and y>0:
        return str(x) + '+' + str(y)
    elif x==0 and y>0 :
        return str(y)
    elif x>0 and y == 0:
        return str(x)
    else:
        return str(0)
concat_colums(0,0)   
```

```python
# Apply function in dataset to add new feature
df['Camera quality index (Front + Rear specs in MP)'] = df.apply(lambda x: concat_colums(x['Front Camera'],x['Rear Camera']),axis = 1)
```

```python
# Make a feature for Phone Storage by adding both memory columns
df['Phone storage(RAM +ROM in GB)'] = df.apply(lambda x: concat_colums(x['RAM'],x['Memory']),axis = 1)
```

# Model Building

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import OneHotEncoder, StandardScaler
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
from sklearn.ensemble import RandomForestRegressor
```

```python
x = encoded_mobile_data.drop('Price',axis = 1)
y = encoded_mobile_data['Price']
```

Split the dataset into training and testing sets


```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
x_train.head()
```

```python
scaler = StandardScaler()
x_train_scaled = scaler.fit_transform(x_train)
x_test_scaled = scaler.transform(x_test)
```

```python
x_train_scaled
```

# First Develop Machine learning Model with linear regression

```python
# Make object of Model
lr_model = LinearRegression()
```

```python
# Fit the training dataset in object of model
lr_model.fit(x_train_scaled,y_train)
```

```python
# Predict the test Data with Model
y_prediction = lr_model.predict(x_test_scaled)
```

```python
# Score of model
lr_model.score(x_test_scaled,y_test)
```

This model is giving negative score so we will check it with another model

```python
# Mean Absolute Error
mae1 = mean_absolute_error(y_test, y_prediction)
mae1
```

```python
# Root Mean Square Score
rmse1 = mean_squared_error(y_test, y_prediction, squared=False)
rmse1
```

```python
# R2 score
r2_ = r2_score(y_test, y_prediction)
r2_
```

# Now Develop Machine learning Model with Random Forest

```python
# Make object of model
rf_model = RandomForestRegressor()
```

```python
# Fit the training dataset in object of model
rf_model.fit(x_train_scaled,y_train)
```

```python
# Predict the test Data with Model
y_pred = rf_model.predict(x_test_scaled)
```

```python
# Score of model
rf_model.score(x_test_scaled,y_test)
```

Evaluate the model

```python
# Mean Absolute Error
mae = mean_absolute_error(y_test, y_pred)
```

```python
# Root Mean Square Score
rmse = mean_squared_error(y_test, y_pred, squared=False)
```
